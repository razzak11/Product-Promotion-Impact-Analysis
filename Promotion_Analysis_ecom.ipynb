### Impact Analysis

The impact of promotion is observed using the Conversion Rate (CR)

Calculating a Baseline CR 

-  To make sure the impact observed is due to promotion only not because of some other factors.

```python
baseline_cr=sales[sales.product_id.isin(promo.ProductId)]
baseline_cr.head()
```

```python
'''
    Here i have calculated cr for each promoted product on its non-promotion days 
    followed by that i calculated the aggregated/avg cr for the same product and that is our baseline cr for that product
    which is then compared with promotion day cr to analyse the impact.
'''

def fun1(x):
    c=0
    for event in x:
        if event=='Purchase':
            c+=1
    return c/len(x)

base_cr_each_product={}    
for p_id in promo.ProductId:
    datee=promo[promo.ProductId==p_id].Date.iloc[0]
    base_cr_each_product[p_id]=baseline_cr[(baseline_cr.event_date != datee) & (baseline_cr.product_id==p_id)].groupby('event_date').agg(CR=('event_type',fun1))['CR'].mean()
```

```python
baseline_cr_final=pd.DataFrame(base_cr_each_product.values(),index=base_cr_each_product.keys()).reset_index()
baseline_cr_final.columns=['product_id','Base_CR']
baseline_cr_final
```

### Promotion day CR for each promoted product

```python
# Conversion Rate of Promoted product on promotion day
def fun1(x):
    c=0
    for event in x:
        if event=='Purchase':
            c+=1
    return c/len(x)        

prod_level_CR=promoted_events.groupby(['event_date','product_id']).agg(CR=('event_type',fun1)).reset_index()
prod_level_CR=prod_level_CR.sort_values(by= 'event_date')
prod_level_CR
```

```python
# calculating revenue from promoted products
def fun(grp):
    grp=grp[grp.event_type=='Purchase']
    return grp.price.sum()
prod_level_rev=promoted_events.groupby(['event_date','product_id']).apply(fun).reset_index()
prod_level_cr_rev=prod_level_CR.merge(prod_level_rev,on=['event_date','product_id'],how='inner')
prod_level_cr_rev.rename(columns={0:'Total Revenue'},inplace=True)
```

```python
prod_level_cr_rev
```

### Comparison

```python
impact_analysis_final=baseline_cr_final.merge(prod_level_cr_rev, how='left', on='product_id')
impact_analysis_final
```

```python
def fun(row):
    return row.unique()[0]

temp=promoted_events.groupby('product_id').agg(Brand=('brand',fun),Category=('category',fun)).reset_index()
impact_analysis_final=temp.merge(impact_analysis_final,how='right',on='product_id')
impact_analysis_final
```

- The nan values of CR are corresponding to those 25 products which couldnt gather customer attention as also mentioned above.


```python
with pd.ExcelWriter('Data_summary.xlsx') as writer:
    impact_analysis_final.to_excel(writer, sheet_name='Sheet1', index=False)
    Event_and_TotalPromo_summary.to_excel(writer, sheet_name='Sheet2', index=False)
    brand_prodcat_wise_interactions.to_excel(writer,sheet_name='sheet3',index=False)
```

```python
cr_final= impact_analysis_final.dropna()
cr_final= cr_final[cr_final['CR']>cr_final['Base_CR']]
cr_final
```

Final Observations based on CR:

-   Out of 61 products promoted 25 did not show any type of traffic i.e no view,cart and purchase event occured.
-   13 (shown above) have shown significant impact.
-   The rest (61-25-13) 23 products had no significant impact as their CR is lower than the base CR.
